### Lets use the strongest correlated attributes to create a MLR Model to predict 'HST' for factors only knowable pre kick-off


#### Check for any missing values

```python
df.isna().sum()
```

#### Remove NaN values

```python
df.dropna(inplace=True)
df.isna().sum()
```

#### Lets assess the correlations again

```python
df.corr()['HST'].sort_values()
```

To avoid multicollinearity with the independent variables we will start with these correlated attributes:
HST_f_Av, HG_Exp, AST_a_Av, HS_f_Av, AS_a_Av, Exp, HST_Lg_Av, AST_Lg_Av, AST_f_Av, HST_a_Av, AS_f_Av, HS_a_Av, AG_Exp

#### Calculate the R_squared value of an MLR model of these attributes 

```python
F = df[['HST_f_Av', 'HG_Exp', 'AST_a_Av', 'HS_f_Av', 'AS_a_Av', 'HST_Lg_Av', 'AST_Lg_Av', 'AST_f_Av', 'HST_a_Av', 'AS_f_Av', 'HS_a_Av', 'AG_Exp']]
A = df['HST']
lm = LinearRegression()
lm.fit(F,A)
print('The R-Square is: ', lm.score(F,A))
print(lm.intercept_)
print(lm.coef_)
```

Initial R2 of 0.323
Next, split the data 80:20 and do a Train/Test analysis to test the R2 on unseen data

```python
#Split our data separating the independent and target variables
y_data = df['HST']
x_data = df.drop('HST', axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=1)
print('number of test samples :', x_test.shape[0])
print('number of training samples :', x_train.shape[0])
```

```python
lr = LinearRegression()
lr.fit(x_train[['HST_f_Av', 'HG_Exp', 'AST_a_Av', 'HS_f_Av', 'AS_a_Av', 'HST_Lg_Av', 'AST_Lg_Av', 'AST_f_Av', 'HST_a_Av', 'AS_f_Av', 'HS_a_Av', 'AG_Exp']],y_train)
x = lr.score(x_test[['HST_f_Av', 'HG_Exp', 'AST_a_Av', 'HS_f_Av', 'AS_a_Av', 'HST_Lg_Av', 'AST_Lg_Av', 'AST_f_Av', 'HST_a_Av', 'AS_f_Av', 'HS_a_Av', 'AG_Exp']],y_test)

print('The R-Square is: ', x)
```

R2 of 0.32 matches our expected. We can assert that 32% of variation can be explained by the model.

#### We will now simplify our model by removing some attributes to assess individual impact on R_Sq

Lets remove team and league 'Shots' averages For and Against and just keep team SOT averages

```python
F = df[['HST_f_Av', 'HG_Exp', 'AST_a_Av', 'HST_Lg_Av', 'AST_Lg_Av', 'AST_f_Av', 'HST_a_Av', 'AG_Exp']]
A = df['HST']
lm = LinearRegression()
lm.fit(F,A)
print('The R-Square is: ', lm.score(F,A))
print(lm.intercept_)
print(lm.coef_)
```

R2 of 0.32 suggests shots (all) does not add any extra value to our model. We will remove those attributes permanently. 
Lets now remove 'Away' factors to assess their impact.

```python
F = df[['HST_f_Av', 'HG_Exp', 'HST_Lg_Av', 'HST_a_Av']]
A = df['HST']
lm = LinearRegression()
lm.fit(F,A)
print('The R-Square is: ', lm.score(F,A))
print(lm.intercept_)
print(lm.coef_)
```

R2 of 0.248 suggests 'away' criteria does have a positive affect on model accuracy. We will add them back in. Next we will remove the league averages home and away SOT to assess their impact.

```python
F = df[['HST_f_Av', 'HG_Exp', 'AST_a_Av', 'AST_f_Av', 'HST_a_Av', 'AG_Exp']]
A = df['HST']
lm = LinearRegression()
lm.fit(F,A)
print('The R-Square is: ', lm.score(F,A))
print(lm.intercept_)
print(lm.coef_)
```

R2 of 0.312. Minor reduction from 0.323 so we will keep the averages in.

Having tested various combinations, the best R2 is attained from using the attributes of 'HST_f_Av', 'HG_Exp', 'AST_a_Av', 'HST_Lg_Av', 'AST_Lg_Av', 'AST_f_Av', 'HST_a_Av', 'AG_Exp'.
 Our finalized model formula for predicting HST is:
##### HST(predicted) = -0.11 + (0.96 * HST_f_Av) + (0.08 * 'HG_Exp') + (0.97 * AST_a_Av) + (-0.94 * HST_Lg_Av) +
##### (-0.06 * AST_Lg_Av) + (-0.01 * AST_f_Av) + (0.01 * HST_a_Av) + (0.13 * AG_Exp)

### Lets do the same for AST

We will keep the same attributes used for the optimum HST R2

```python
F = df[['HST_f_Av', 'HG_Exp', 'AST_a_Av', 'HST_Lg_Av', 'AST_Lg_Av', 'AST_f_Av', 'HST_a_Av', 'AG_Exp']]
A = df['AST']
lm = LinearRegression()
lm.fit(F,A)
print('The R-Square is: ', lm.score(F,A))
print(lm.intercept_)
print(lm.coef_)
```

The non trivial coefficients generated suggest that they all have a worthwhile impact on R2, therefore we will use the same attributes for AST as HST in our final AST model.


 Our finalized regression model formula for predicting AST therefore is:
##### AST (predicted) = -0.21 + (-0.01 * HST_f_Av) + (0.18 * 'HG_Exp') + (0.01 * AST_a_Av) + (-0.03 * HST_Lg_Av) +
##### (-0.99 * AST_Lg_Av) + (0.97 * AST_f_Av) + (0.98 * HST_a_Av) + (0.16 * AG_Exp)
