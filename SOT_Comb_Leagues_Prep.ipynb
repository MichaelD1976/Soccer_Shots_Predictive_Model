```python
def add_HST_AST_for_against_columns(df):
    
    df['HS_f_Av'] = round(df.groupby('HomeTeam')['HS'].transform('mean'),2)
    df['HS_a_Av'] = round(df.groupby('HomeTeam')['AS'].transform('mean'),2)
    df['AS_f_Av'] = round(df.groupby('AwayTeam')['AS'].transform('mean'),2)
    df['AS_a_Av'] = round(df.groupby('AwayTeam')['HS'].transform('mean'),2)   
    
    df['HST_f_Av'] = round(df.groupby('HomeTeam')['HST'].transform('mean'),2)
    df['HST_a_Av'] = round(df.groupby('HomeTeam')['AST'].transform('mean'),2)
    df['AST_f_Av'] = round(df.groupby('AwayTeam')['AST'].transform('mean'),2)
    df['AST_a_Av'] = round(df.groupby('AwayTeam')['HST'].transform('mean'),2)
    
    df['HST_Lg_Av'] = round(df['HST'].mean(),2)
    df['AST_Lg_Av'] = round(df['AST'].mean(),2)

    df['HS_Lg_Av'] = round(df['HS'].mean(),2)
    df['AS_Lg_Av'] = round(df['AS'].mean(),2)    
    
    return df

df1 = add_HST_AST_for_against_columns(df1)
df2 = add_HST_AST_for_against_columns(df2)
df3 = add_HST_AST_for_against_columns(df3)
df4 = add_HST_AST_for_against_columns(df4)
df5 = add_HST_AST_for_against_columns(df5)
df6 = add_HST_AST_for_against_columns(df6)
df7 = add_HST_AST_for_against_columns(df7)
df8 = add_HST_AST_for_against_columns(df8)
df9 = add_HST_AST_for_against_columns(df9)
df10 = add_HST_AST_for_against_columns(df10)
df11 = add_HST_AST_for_against_columns(df11)
df12 = add_HST_AST_for_against_columns(df12)
df13 = add_HST_AST_for_against_columns(df13)
df14 = add_HST_AST_for_against_columns(df14)
df15 = add_HST_AST_for_against_columns(df15)
df16 = add_HST_AST_for_against_columns(df16)
df17 = add_HST_AST_for_against_columns(df17)
df18 = add_HST_AST_for_against_columns(df18)
df19 = add_HST_AST_for_against_columns(df19)
df20 = add_HST_AST_for_against_columns(df20)
df21 = add_HST_AST_for_against_columns(df21)
df22 = add_HST_AST_for_against_columns(df22)
df23 = add_HST_AST_for_against_columns(df23)
#df24 = add_HST_AST_for_against_columns(df24)
df25 = add_HST_AST_for_against_columns(df25)
df26 = add_HST_AST_for_against_columns(df26)
df27 = add_HST_AST_for_against_columns(df27)
df28 = add_HST_AST_for_against_columns(df28)
df29 = add_HST_AST_for_against_columns(df29)
df30 = add_HST_AST_for_against_columns(df30)
df31 = add_HST_AST_for_against_columns(df31)
df32 = add_HST_AST_for_against_columns(df32)
df33 = add_HST_AST_for_against_columns(df33)
df34 = add_HST_AST_for_against_columns(df34)
df35 = add_HST_AST_for_against_columns(df35)
df36 = add_HST_AST_for_against_columns(df36)
df37 = add_HST_AST_for_against_columns(df37)
df38 = add_HST_AST_for_against_columns(df38)
df39 = add_HST_AST_for_against_columns(df39)
df40 = add_HST_AST_for_against_columns(df40)
df41 = add_HST_AST_for_against_columns(df41)
df42 = add_HST_AST_for_against_columns(df42)
df43 = add_HST_AST_for_against_columns(df43)
df44 = add_HST_AST_for_against_columns(df44)
df45 = add_HST_AST_for_against_columns(df45)
df46 = add_HST_AST_for_against_columns(df46)
df47 = add_HST_AST_for_against_columns(df47)
df48 = add_HST_AST_for_against_columns(df48)
df49 = add_HST_AST_for_against_columns(df49)
df50 = add_HST_AST_for_against_columns(df50)
df51 = add_HST_AST_for_against_columns(df51)
df52 = add_HST_AST_for_against_columns(df52)
df53 = add_HST_AST_for_against_columns(df53)
df54 = add_HST_AST_for_against_columns(df54)
df55 = add_HST_AST_for_against_columns(df55)
df56 = add_HST_AST_for_against_columns(df56)
df57 = add_HST_AST_for_against_columns(df57)
df58 = add_HST_AST_for_against_columns(df58)
df59 = add_HST_AST_for_against_columns(df59)
df60 = add_HST_AST_for_against_columns(df60)

#for df in (df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15,df16,df17,df18,df19,df20):
#    df = add_HST_AST_for_against_columns(df)

```

##### If 'Avg' column NaN take value from  'BbAv'

```python
df['AvH'] = df['AvgH'].fillna(df['BbAvH'])
```

```python
df['AvD'] = df['AvgD'].fillna(df['BbAvD'])
```

```python
df['AvA'] = df['AvgA'].fillna(df['BbAvA'])
```

```python
df['Av_O2.5'] = df['Avg>2.5'].fillna(df['BbAv>2.5'])
df['Av_U2.5'] = df['Avg<2.5'].fillna(df['BbAv<2.5'])
```

```python
df = df.drop(['AvgH','AvgD','AvgA','Avg>2.5', 'Avg<2.5','BbAvH', 'BbAvD', 'BbAvA', 'BbAv>2.5','BbAv<2.5'], axis=1)
```

#### Add columns with Totals (Goals, Shots)

```python
df['TG'] = df['FTHG'] + df['FTAG']  # adding a Total Goals column
```

```python
df['TST'] = df['HST'] + df['AST']  # adding a Total Shots on Target column
```

```python
df.columns
```

#### Add column with Home/Draw/Away Win % & Ov/Un (True)

```python
df['HW_%'] = round((1/df['AvH']) / (1/df['AvH'] + 1/df['AvD'] + 1/df['AvA']),2)  # Add a column giving the Home Win % (HW_%)
```

```python
df['X_%'] = round((1/df['AvD']) / (1/df['AvH'] + 1/df['AvD'] + 1/df['AvA']),2)  # Add a column giving the Draw % (X_%)
```

```python
df['AW_%'] = round((1/df['AvA']) / (1/df['AvH'] + 1/df['AvD'] + 1/df['AvA']),2)  # Add a column giving the Away Win % (AW_%)
```

```python
df['Ov2.5_%'] = round((1/df['Av_O2.5']) / (1/df['Av_U2.5'] + 1/df['Av_O2.5']),2)  # Add a column giving the Over 2.5 % (Un2.5_%)
```

```python
df['Un2.5_%'] = round((1/df['Av_U2.5']) / (1/df['Av_O2.5'] + 1/df['Av_U2.5']),2)  # Add a column giving the Under 2.5 % (Un2.5_%)
```

```python
dfOU = pd.read_csv('C:/Users/MikeD/Documents/Betting/OverUnder_Exp.csv', index_col=False)
dfOU
```

##### Add Home/Away Expected Goals

```python
df['HG_Exp'] = round((df['Exp']) * (df['HW_%'] + 0.5 * df['X_%']),2)
df['AG_Exp'] = round((df['Exp']) * (df['AW_%'] + 0.5 * df['X_%']),2)
df['Sup Exp'] = df['HG_Exp'] - df['AG_Exp']
df
```

```python
# drop bookie prices now we can trues collated
df = df.drop(['AvH','AvD','AvA','Av_O2.5', 'Av_U2.5'], axis=1)
df
```
